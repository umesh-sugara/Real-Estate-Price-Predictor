## *Scikit-Learn Design*

primarilary 3 types of objects:
1. Estimators -- Estimates the mising values (Imputers)
                It has a fit method that fits the dataset and calculates internal parameters

2. Transformers -- Takes input and return output based on the learning of Fit Method.
        It has a convinience function FIT-TRANSFORM . if fits and then transforms. it is optimised forms of FIT+Transform. better to use it.

3. Predictors-- Regressors. It have 2 common functions.
                FIT-- it fits the dataset to calculate the parameters.
                PREDICT()---it will predict the values and also score to evaluate the predictions
